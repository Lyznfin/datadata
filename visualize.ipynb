```python
import csv
import matplotlib.pyplot as plt

# Initialize lists to store data
occurrences = []
artist_and_category = []

# Read data from CSV file
with open('final_sorted_count.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        occurrences.append(int(row['occurrences']))
        artist_and_category.append(f"{row['artist']}")
        if occurrences.__len__() > 9:
            break

# Create plot
plt.figure(figsize=(22, 18))
plt.barh(artist_and_category, occurrences, color='green')
plt.xlabel('Jumlah streaming lagu')
plt.ylabel('Artist dan genre')
plt.title('Top 10 artist dengan pendengar terbanyak')
plt.gca().invert_yaxis()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
from random import randrange

genre_data = {}
with open('top_5_sorted_output_by_genre.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        genre = row['genre']
        if genre not in genre_data:
            genre_data[genre] = {'artists': [], 'occurrences': []}
        genre_data[genre]['artists'].append(row['artist'])
        genre_data[genre]['occurrences'].append(int(row['occurrences']))

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', 
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf', 
          '#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

for genre, data in genre_data.items():
    plt.figure(figsize=(12, 8))
    plt.barh(data['artists'], data['occurrences'], color=colors[randrange(0, colors.__len__() - 1)])
    plt.xlabel('Number of Streams')
    plt.ylabel('Artist')
    plt.title(f'Top 5 Artists in {genre} genre')
    plt.gca().invert_yaxis()
    plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
from random import randrange

genre_data = {}
with open('top_5_sorted_output_by_genre.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        genre = row['genre']
        if genre not in genre_data:
            genre_data[genre] = {'artists': [], 'occurrences': []}
        genre_data[genre]['artists'].append(row['artist'])
        genre_data[genre]['occurrences'].append(int(row['occurrences']))

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', 
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf', 
          '#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

for genre, data in genre_data.items():
    plt.figure(figsize=(12, 12))
    plt.pie(data['occurrences'], labels=data['artists'], autopct='%1.1f%%', startangle=140)
    plt.axis('equal')
    plt.title(f'Top 5 Artists in {genre} genre')
    plt.show()
```
